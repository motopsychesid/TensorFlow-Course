## Preprocess data

### Subtask:
Clean and preprocess the email data for both spam and fraud detection, including handling missing values, tokenization, and feature extraction.

**Reasoning**:

Before training any machine learning models, it is essential to preprocess the data. This involves several steps:

1.  **Handling Missing Values**: Identify and address any missing values in the relevant columns of both dataframes.
2.  **Text Preprocessing**: Clean the text data by removing irrelevant characters, converting text to lowercase, and performing tokenization.
3.  **Feature Extraction**: Convert the text data into numerical features that can be used by machine learning models. For text data, TF-IDF (Term Frequency-Inverse Document Frequency) is a common and effective technique.

```python
# Preprocessing for Spam Detection (using df_emails)

# Handle missing values in the 'Prediction' column (if any) - though info shows no missing values
# For simplicity in this example, we assume 'Prediction' is the target variable and other columns are features.
# If there were missing values in feature columns, we would need to impute or drop them.

# The df_emails dataset appears to be already preprocessed into numerical features (word counts).
# We will assume these are the features and 'Prediction' is the target.
X_spam = df_emails.drop(['Prediction', 'Email No.'], axis=1)
y_spam = df_emails['Prediction']

print("Spam Detection Features (X_spam) Head:")
display(X_spam.head())
print("\nSpam Detection Target (y_spam) Head:")
display(y_spam.head())


# Preprocessing for Fraud Detection (using df_fraud)

# Handle missing values in relevant columns
# We'll focus on the 'Body' column for text-based features and 'Label' for the target.
df_fraud.dropna(subset=['Body', 'Label'], inplace=True)

# Text Preprocessing (Tokenization and cleaning)
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize


stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text) # Remove non-alphabetic characters
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(tokens)

df_fraud['Clean_Body'] = df_fraud['Body'].apply(preprocess_text)

# Feature Extraction (TF-IDF)
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(max_features=5000) # Limit to top 5000 features

X_fraud_text_features = tfidf_vectorizer.fit_transform(df_fraud['Clean_Body']).toarray()
X_fraud = pd.DataFrame(X_fraud_text_features)
y_fraud = df_fraud['Label']

print("\nFraud Detection Features (X_fraud) Head:")
display(X_fraud.head())
print("\nFraud Detection Target (y_fraud) Head:")
display(y_fraud.head())
```

## Train models

### Subtask:
Train separate machine learning models for spam detection and fraud detection using the training data.

**Reasoning**:
Now that the data is preprocessed and split, we can train our machine learning models. For spam detection, the dataset is already in a numerical format suitable for various classifiers. Given the nature of text data and the success of Naive Bayes in text classification, we will use a Multinomial Naive Bayes classifier for spam detection.

LightGBM is a gradient boosting framework that is known for its speed and efficiency, and it can handle large datasets with many features effectively.

```python
from sklearn.naive_bayes import MultinomialNB
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Train Multinomial Naive Bayes model for Spam Detection
mnb_spam_model = MultinomialNB()
mnb_spam_model.fit(X_spam_train, y_spam_train)

# Split data for Fraud Detection and Train LightGBM model
X_fraud_train, X_fraud_test, y_fraud_train, y_fraud_test = train_test_split(X_fraud, y_fraud, test_size=0.2, random_state=42)

lgb_fraud_model = lgb.LGBMClassifier(random_state=42)
lgb_fraud_model.fit(X_fraud_train.values.astype(float), y_fraud_train.values.astype(int))

print("Spam detection model training complete.")
print("Fraud detection model training complete.")
```

## Preprocess, Split, and Train Models

### Subtask:
Combine the preprocessing, data splitting, and model training steps for both spam and fraud detection into a single cell to ensure variable availability.

**Reasoning**:
To avoid `NameError`s and ensure all necessary variables are defined and available for subsequent steps, we will consolidate the preprocessing, data splitting, and model training for both spam and fraud detection into one code cell. This guarantees that the preprocessed data and the resulting training and testing sets are immediately used for training the models within the same execution context.

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
import lightgbm as lgb
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# NLTK data download check is now handled in a separate cell (c01df18d)
# Assuming NLTK data is downloaded and available


# Load Data (assuming data is already loaded into df_emails and df_fraud from a previous cell)
# If not, uncomment the following lines:
# try:
#     df_emails = pd.read_csv('/content/emails.csv', engine='python')
#     print("Emails DataFrame loaded successfully.")
# except Exception as e:
#     print(f"Error loading emails.csv: {e}")

# try:
#     df_fraud = pd.read_csv('/content/enron_data_fraud_labeled.csv')
#     print("Fraud DataFrame loaded successfully.")
# except Exception as e:
#     print(f"Error loading enron_data_fraud_labeled.csv: {e}")


# --- Spam Detection ---

# Preprocessing for Spam Detection (using df_emails)
X_spam = df_emails.drop(['Prediction', 'Email No.'], axis=1)
y_spam = df_emails['Prediction']

# Split data for Spam Detection
X_spam_train, X_spam_test, y_spam_train, y_spam_test = train_test_split(X_spam, y_spam, test_size=0.2, random_state=42)

# Train Multinomial Naive Bayes model for Spam Detection
mnb_spam_model = MultinomialNB()
mnb_spam_model.fit(X_spam_train, y_spam_train)

print("\nSpam detection preprocessing, splitting, and model training complete.")
print("X_spam_train shape:", X_spam_train.shape)
print("y_spam_train shape:", y_spam_train.shape)


# --- Fraud Detection ---

# Preprocessing for Fraud Detection (using df_fraud)

# Handle missing values in relevant columns
df_fraud.dropna(subset=['Body', 'Label'], inplace=True)

# Text Preprocessing (Tokenization and cleaning)
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text) # Remove non-alphabetic characters
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(tokens)

df_fraud['Clean_Body'] = df_fraud['Body'].apply(preprocess_text)

# Feature Extraction (TF-IDF)
tfidf_vectorizer = TfidfVectorizer(max_features=5000) # Limit to top 5000 features

X_fraud_text_features = tfidf_vectorizer.fit_transform(df_fraud['Clean_Body']).toarray()
X_fraud = pd.DataFrame(X_fraud_text_features)
y_fraud = df_fraud['Label']

# Split data for Fraud Detection
X_fraud_train, X_fraud_test, y_fraud_train, y_fraud_test = train_test_split(X_fraud, y_fraud, test_size=0.2, random_state=42)

# Train LightGBM model for Fraud Detection
lgb_fraud_model = lgb.LGBMClassifier(random_state=42)
lgb_fraud_model.fit(X_fraud_train.values.astype(float), y_fraud_train.values.astype(int))


print("\nFraud detection preprocessing, splitting, and model training complete.")
print("X_fraud_train shape:", X_fraud_train.shape)
print("y_fraud_train shape:", y_fraud_train.shape)
```

## Evaluate models

### Subtask:
Evaluate the performance of the trained spam detection model using appropriate metrics and visualize the results.

**Reasoning**:
Now that the spam detection model is trained, we need to evaluate its performance on the unseen test data. We will use common classification metrics such as Accuracy, Precision, Recall, and F1-score. Additionally, we will generate a confusion matrix to visualize the model's performance in terms of true positives, true negatives, false positives, and false negatives.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Evaluate Spam Detection Model
y_spam_pred = mnb_spam_model.predict(X_spam_test)

spam_accuracy = accuracy_score(y_spam_test, y_spam_pred)
spam_precision = precision_score(y_spam_test, y_spam_pred)
spam_recall = recall_score(y_spam_test, y_spam_pred)
spam_f1 = f1_score(y_spam_test, y_spam_pred)

print("Spam Detection Model Evaluation:")
print(f"Accuracy: {spam_accuracy:.4f}")
print(f"Precision: {spam_precision:.4f}")
print(f"Recall: {spam_recall:.4f}")
print(f"F1-score: {spam_f1:.4f}")

# Visualize Confusion Matrix for Spam Detection
cm_spam = confusion_matrix(y_spam_test, y_spam_pred)
disp_spam = ConfusionMatrixDisplay(confusion_matrix=cm_spam, display_labels=['Not Spam', 'Spam'])
disp_spam.plot()
plt.title('Spam Detection Confusion Matrix')
plt.show()
```

## Classify Input Email

### Subtask:
Preprocess the input email text and classify it as spam or not spam using the trained spam detection model.

**Reasoning**:
To classify the new email using the trained Multinomial Naive Bayes model, we need to apply the same preprocessing steps that were used for the training data. This involves converting the text into a numerical feature vector based on the vocabulary learned from the original `emails.csv` dataset. Since the original dataset contains word counts, we will create a feature vector for the input email that represents the counts of words present in the training vocabulary.

```python
# Input email text provided by the user
input_email_text = """
[Required] Email Verification
Inbox

foundit <updates@alerts.foundit.in> Unsubscribe
Fri, Jul 25, 3:47 PM
to me


Dear B. Yuvan Siddarth,

If your email isn’t current, you might not receive updates about your profile or actions you’ve taken recently.

We recommend verifying now to avoid missing future messages that could help you move forward.

Verify Email
If you'd like to unsubscribe and stop receiving these emails click here .
"""

# Preprocess the input email text to get word counts based on the training vocabulary
# We need to use the same vocabulary as the training data (X_spam)

# Get the list of feature names (words) from the training data
feature_names = X_spam.columns.tolist()

# Create a dictionary to store word counts for the input email
input_word_counts = {feature: 0 for feature in feature_names}

# Simple tokenization and counting (assuming basic space-based splitting and lowercasing)
# A more robust approach would use the same tokenizer as used for the original dataset if available
cleaned_text = input_email_text.lower()
words = cleaned_text.split()

for word in words:
    # Remove non-alphabetic characters and check if the word is in the training vocabulary
    cleaned_word = re.sub(r'[^a-zA-Z]', '', word)
    if cleaned_word in input_word_counts:
        input_word_counts[cleaned_word] += 1

# Convert the word counts to a pandas DataFrame in the same order as X_spam
X_input_email = pd.DataFrame([input_word_counts])

# Ensure the columns are in the same order as the training data
X_input_email = X_input_email[feature_names]


# Predict using the trained Multinomial Naive Bayes model
spam_prediction = mnb_spam_model.predict(X_input_email)

# Output the result
if spam_prediction[0] == 1:
    print("The email is classified as: SPAM")
else:
    print("The email is classified as: NOT SPAM")
```
